### 2. 데이터 탐색

2-1 데이터 컬럼명, 구성 등 대략적인 파악
- df.head(), df.tail() : 데이터 걸럼명등등 파악
- df.shape : DataFrame 크기(행, 열)파악
- df.describe() : 기술통계량 파악 (데이터 수, 평균, 표준편차, 최소값, 1~3사분위수 등등)

--> df.unique(), df.value_counts(),df.columns 등등으로 데이터 파악
