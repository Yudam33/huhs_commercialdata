## 목표 1. '중점관리 키워드'와 '저효율 키워드' 추출

###   - 중점 관리 키워드 
 1. 대중적인 키워드 中 중점 관리할 키워드 : '노출수'가 상위 10%인 값 추출 -> '클릭률%'로 내림차순 정렬 - > 상위 30개 추출
 2. 다소 마이너틱하나, 구매로 이어질 확률이 높아 중점 관리할 키워드 : '클릭률%'가 상위 5%인 값 추출 -> '클릭수'로 내림차순 정렬 -> 상위 10개 추출
 
### - 저효율 키워드
1. 대중성있으나 클릭률이 지나치게 낮은 키워드: '노출수'가 상위 50%인 값 추출 -> '클릭률%' 오름차순 정렬 -> 상위 30개 추출
2. 노출 자체가 되지 않는 키워드: '노출수'가 하위 10%인 값 추출 -> '클릭률%'로 오름차순 정렬 -> 상위 30개 추출

----------------------------------------------------
* 사용자 접근성만을 지표로 하며, 광고 비용은 고려하지 않는다.
* 여러 광고그룹에서 중복 사용하고 있는 키워드는 노출수 / count 후 반영
